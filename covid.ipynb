```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://data.cdc.gov/api/views/9mfq-cb36/rows.csv?accessType=DOWNLOAD')
df.head()
```

```python
# Choropleth Map of 7-Day Moving Average of New Cases - Animated

import plotly.graph_objs as go
choropleth_map = go.Figure(
    data = {
        'type':'choropleth',
        'locations':df1['state'],
        'locationmode':'USA-states',        
        'colorscale':'YlGnBu',            
        'text':df1['state'],
        'z':df1['7D_cases_mean_per_100K'],
        'colorbar':{'title':'New Cases per 100K'}
    },     
    layout = {
      'geo':{
          'scope':'usa'
      }  
    })

choropleth_map
```
